### Scrape data from Wikipedia into a .CSV file

```python
import pandas as pd
import wikipedia as wp
 
#Get the html source
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
df = pd.read_html(html)[0]
df.to_csv('toronto.csv',header=0,index=False)
print (df)
```
